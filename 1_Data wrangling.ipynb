After overviewing the data it is apparent that the policy code labeling changed over time. Prior the year 2016 the rejected loan applications were labeled as `0` and the after that it was changed to `2`. Keeping that in mind I will leave all the values and only change "2" to "0" in the policy code column.

```python
rejected["policy_code"] = rejected["policy_code"].replace(2, 0)
```

The accepted loan request dataset doesn't have a "risk_score" feature. To calculate this feature I will use the "fico_range_high" and "fico_range_low" features and calculate the mean of those two values.

```python
accepted["risk_score"] = ((accepted["fico_range_high"] + accepted["fico_range_low"]) / 2).astype(float)
```

After droping the rows with all missing values we can see that my target features don't contain null values anymore. But now I need to do some feature engineering to finalize my dataset preparation. The first feature is `term` from which I will extract the numeric part. Also I will create a new feature called `time_since_earliest_cr_line` in which I will be calculating the time in days since the first clients credit line was opened.

```python
accepted["term"] = accepted["term"].str.extract(r"(\d+)", expand=False).astype("Int32")
```

```python
accepted["earliest_cr_line"] = pd.to_datetime(accepted["earliest_cr_line"], format="%b-%Y")
accepted["issue_d"] = pd.to_datetime(accepted["issue_d"], format="%b-%Y")
```

```python
accepted["time_since_earliest_cr_line"] = (accepted["issue_d"] - accepted["earliest_cr_line"]).dt.days
```
